Create table nmbr_events.  State, total number of events

```python

qry = 'SELECT STATE, count(STATE) as NMBR_EVENTS \
FROM all_events \
GROUP BY STATE;'

nmbr_events = pd.read_sql(qry, conn)

nmbr_events.head()
```

```python
# Create/load to db
nmbr_events.to_sql('nmbr_events', conn, if_exists='replace')
```

```python
# DEATHS_DIRECT
noaa_cats_noZero =noaa_cats.loc[(noaa_cats["DEATHS_DIRECT"]  != 0 ) | (noaa_cats["DEATHS_INDIRECT"]  != 0 )]
```

# sum_state_death table

```python
sumStateDeath = noaa_cats_noZero.groupby(['STATE']).sum()
sumStateDeath.columns
```

```python
sumStateDeath['TOTAL_DEATHS'] = sumStateDeath['DEATHS_DIRECT'] + sumStateDeath['DEATHS_INDIRECT']
```

```python
sumStateDeath.head()
```

```python
sumStateDeath.reindex()
# sumStateDeath.head()
```

```python
sumStateDeath.to_sql('sum_state_death', conn, if_exists='replace')
```

# sum_state_event table

```python
sumEventState = noaa_cats_noZero.groupby(['STATE', 'EVENT_TYPE']).sum()
```

```python
sumEventState.columns
```

```python
sumEventState['TOTAL_DEATHS'] = sumEventState['DEATHS_DIRECT'] + sumEventState['DEATHS_INDIRECT']
```

```python
sumEventState.to_sql('sum_state_event', conn, if_exists='replace')
```

# sum_state_event_year table

```python
# now do a table group by state/event/year so we have a time sequence
state_event_year = noaa_cats_noZero.groupby(['STATE', 'EVENT_TYPE', 'YEAR']).sum()
```

```python
state_event_year.columns
```

```python
state_event_year['TOTAL_DEATHS'] = state_event_year['DEATHS_DIRECT'] + state_event_year['DEATHS_INDIRECT']
```

```python
state_event_year.to_sql('sum_state_event_year', conn, if_exists='replace')
```

```python
# sum_state_event table
qry = 'SELECT * FROM sum_state_event as s \
JOIN state_abbrv as a \
ON s.STATE = a.STATE_TITLE;'
sumsevnt_df = pd.read_sql(qry, conn)

# index STATE_CAPS
sumsevnt_df.drop(columns= ['index', 'STATE_CAPS', 'STATE_TITLE', 'STATE_APSTYLE'], inplace=True)
sumsevnt_df.to_sql('sum_state_event', conn, if_exists='replace')
```

```python
# sum_state_event_year table
qry = 'SELECT * FROM sum_state_event_year as s \
JOIN state_abbrv as a \
ON s.STATE = a.STATE_TITLE;'
sum_state_event_year_df = pd.read_sql(qry, conn)

# index STATE_CAPS
sum_state_event_year_df.drop(columns= ['index', 'STATE_CAPS', 'STATE_TITLE', 'STATE_APSTYLE'], inplace=True)
sum_state_event_year_df.to_sql('sum_state_event_year', conn, if_exists='replace')
```

```python
# nmbr_events table
qry = 'SELECT * FROM nmbr_events as s \
JOIN state_abbrv as a \
ON s.STATE = a.STATE_TITLE;'
nmbr_events_df = pd.read_sql(qry, conn)

# index STATE_CAPS
nmbr_events_df.drop(columns= ['index', 'STATE_CAPS', 'STATE_TITLE', 'STATE_APSTYLE'], inplace=True)
nmbr_events_df.to_sql('nmbr_events', conn, if_exists='replace')
```
